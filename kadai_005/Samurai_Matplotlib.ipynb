```python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3,tight_layout=True)

#折れ線
y0 = [1, 2, -2, 0]
#プロット
axes[0,0].plot(y0)


#sin関数
x1 = np.linspace(0, 10, 100)
y1 = 2 + 2 * np.sin(2 * x1)
# プロット
axes[0,1].plot(x1, y1, linewidth=2.0)


#ヒストグラム
x2 = np.random .normal(15, 5, 1000)
# プロット
axes[0,2].hist(x2, bins=30)


#散布図
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 100)
y3 = 4 + np.random.normal(0, 2, len(x3))
# プロット
axes[1,0].scatter(x3, y3)


#出力
plt.show()
```
