```python
import matplotlib.pyplot as plt
import japanize_matplotlib
import pandas as pd

df = pd.read_csv("sample_pandas_6.csv")
category_df = pd.read_csv('category.csv')
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

y=df['カテゴリー'].value_counts()
print(df['カテゴリー'].value_counts())

y.plot(kind='bar')
plt.title('カテゴリーごとの注文数')
plt.xlabel('カテゴリー')
plt.ylabel('注文回数')

plt.show()

display(df)

aver=df.groupby('商品番号')['注文数'].mean()

toukei=df.groupby('発注日')['注文数'].describe()

print(aver)
print(toukei)

```
